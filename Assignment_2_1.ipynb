#2. Data Cleaning and Preprocessing

2.1 Remove the first column or Column 1 from the dataset. Treat the year variable as nominal.

```python
# Write your code here

df = df.drop(columns=["Column 1"])

df["year"] = df["year"].astype("category")

print(df.dtypes)
```

2.2 Check for duplicate values and remove them.

```python
# Write your code here

duplicate_count = df.duplicated().sum()
print("Duplicate rows before removal:", duplicate_count)

df = df.drop_duplicates()

duplicate_count_after = df.duplicated().sum()
print("Duplicate rows after removal:", duplicate_count_after)

print("New dataset shape:", df.shape)
```

2.3 Check for missing values. If a data record (row) only has a few missing values, replace the missing values with the median of the column feature in that specific Region variable. If most column values in a data record are missing, remove the data record.

```python
# Write your code here

row_missing_ratio = df.isnull().mean(axis=1)

df = df[row_missing_ratio < 0.5]

print("Shape after removing rows with excessive missing values:", df.shape)

numeric_cols = df.select_dtypes(include=["float64", "int64"]).columns

for col in numeric_cols:
    df[col] = df.groupby("region")[col].transform(
        lambda x: x.fillna(x.median())
        )

print("\nRemaining missing values per column:")
print(df.isnull().sum())


# Any rows that had more than 50% missing values were purged.
# Remaining missing values had numeric features imputed on them using the median within each region.
# Final tally, 18,251 rows and 13 columns with absolutely 0 missing values.
```
